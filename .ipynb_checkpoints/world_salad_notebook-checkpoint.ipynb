## Filtering for Listings With Pay Information

```python
state_abbreviations = [
    "AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DE", "FL", "GA",
    "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD",
    "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ",
    "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC",
    "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"
]
dc_abbreviation = ["DC"]
us = ["United States"]
american_places = state_abbreviations + dc_abbreviation + us

df_in_us = df[df['location'].str.contains('|'.join(state_abbreviations), case=False, na=False)]
```

```python
df_in_us['has_salary_info'] = df_in_us[['max_salary', 'med_salary', 'min_salary', 'pay_period']].notnull().any(axis=1)
df_in_us['has_salary_info'].value_counts()
```

```python
with_salary = df_in_us.loc[df_in_us['has_salary_info'] == True]
with_salary = with_salary.loc[with_salary['currency'] == 'USD']
with_salary.shape
```

```python
# The dataset is big enough that, to work with descriptions, it's more time-economical to
# work with a random sample.
random_sample = with_salary.sample(n=30_000, random_state=42)
# random_sample['preprocessed_description'] = random_sample['description'].apply(preprocess)
```

## Description Preprocessing

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess(text):
    tokens = word_tokenize(text.lower())
    tokens = [word for word in tokens if word not in stop_words]
    tokens = [word for word in tokens if word.isalpha()]
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return ' '.join(lemmatized_tokens)
    
    # if type(text) != str: return ''
    # if text == '': return ''
    # tokens = nltk.word_tokenize(text.lower())
    # tokens = [lemmatizer.lemmatize(token) for token in tokens if token not in stopwords.words('english')]
    # return ' '.join(tokens)
```

```python
random_sample['title_and_description'] = random_sample['title'] + ' ' + random_sample['description']
random_sample['preprocessed_text'] = random_sample['title_and_description'].apply(preprocess)
random_sample.head()
```

```python
# I boil down maximums, minimums and mediums to a single pay number, estimating what the role
# pays in a general sense.

def estimate_pay(row):
    # If there's a medium salary, pick that.
    if pd.notnull(row['med_salary']): return row['med_salary']
    # If there's only a maximum salary, pick that.
    if pd.notnull(row['max_salary']) & pd.isnull(row['min_salary']): return row['max_salary']
    # If there's only a minimum salary, pick that.
    if pd.notnull(row['min_salary']) & pd.isnull(row['max_salary']): return row['min_salary']
    # If there's a maximum and a minimum with no medium, pick the average.
    if pd.notnull(row['max_salary']) & pd.notnull(row['min_salary']):
        return row['max_salary'] + row['min_salary'] / 2
    # If all the useful numbers are missing, return a row we expect to be null.
    return row['max_salary']
```

```python
random_sample['estimated_pay_raw'] = random_sample.apply(estimate_pay, axis=1)
    
random_sample.head()
```

```python
# Pay is described in different pay periods, so the associated number has to be multiplied
# to standardize the pay period. I've selected annual pay as a standard.
def annualize_pay(row):
    if row['pay_period'] == 'YEARLY': return row['estimated_pay_raw']
    if row['pay_period'] == 'MONTHLY': return row['estimated_pay_raw'] * 12
    if row['pay_period'] == 'BIWEEKLY': return row['estimated_pay_raw'] * 26
    if row['pay_period'] == 'WEEKLY': return row['estimated_pay_raw'] * 52
    if row['pay_period'] == 'HOURLY': return row['estimated_pay_raw'] * 2_000
    return row['estimated_pay_raw']
```

```python
random_sample['annualized_pay'] = random_sample.apply(annualize_pay, axis=1)
random_sample.head()
```

## Filtering Outliers

Some listings have extraordinarily high or low pay, which is probably due to an error in the listing being multiplied by the preprocessing. The should be filtered out.

```python
no_outliers = random_sample.loc[random_sample['annualized_pay'] < 500_000]
no_outliers = no_outliers.loc[no_outliers['annualized_pay'] > 500]

no_outliers.shape
```

# Modeling

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(no_outliers['preprocessed_text'])
X_train, X_test, y_train, y_test = train_test_split(X, no_outliers['annualized_pay'], test_size=0.2, random_state=42)
```

## Ridge Model

```python
ridge = Ridge()
ridge.fit(X_train, y_train)

print(ridge.score(X_test, y_test))
```
